```r
# Reading datasets

ny = read.csv('new_york_city.csv')
wash = read.csv('washington.csv')
chi = read.csv('chicago.csv')

library(ggplot2)
```

### Question 1

Find total travel duration for each city.

```r
# Total trip duration for all three dataset.
# Final total duration in seconds duration in datasets is in seconds.

ny_total_duration = sum(ny$Trip.Duration, na.rm=TRUE)
chi_total_duration = sum(chi$Trip.Duration, na.rm=TRUE)
wash_total_duration = sum(wash$Trip.Duration, na.rm=TRUE)

ny_total_duration
chi_total_duration
wash_total_duration
```

```r
# New dataframe created for cities and total duration to plot.

cities = c('new york', 'chicago', 'washington')
total_duration = c(ny_total_duration, chi_total_duration, wash_total_duration)

cities_duration = data.frame(cities, total_duration)

cities_duration
```

```r
# Plotting total duration for each city.
# Plot contains title, x label, y label.

ggplot2::qplot(x=cities, y=total_duration/1000000, data=cities_duration, main='Total travel duration per city',xlab='Cities', ylab='duration in seconds x 10^6')
```

```r
# Numeric Summary that supports the plot

cities_duration
```

**Summary of your question 1 results goes here.**

According to the plot, Washington has the maximum travel duration, followed by New York, followed by Chicago.

### Question 2

**Your question 2 goes here.**

**Find the most common month of starting of travels.**


```r
# Extracting month from datetime for each city and adding it as a new column in the dataframe.

ny$Month <- format(as.Date(ny$Start.Time), "%m")
chi$Month <- format(as.Date(chi$Start.Time), "%m")
wash$Month <- format(as.Date(wash$Start.Time), "%m")
```

```r
# Displaying few rows of New York dataset.
# Notice the las "Month" column.

head(ny)
```

```r
library(ggplot2)
```

```r
# Plotting month wise travels for New York.

ggplot(aes(x=as.numeric(Month)), data=ny) + 
        geom_histogram(binwidth=0.5) +
        scale_x_continuous(breaks=1:6) +
        ggtitle('Histogram of travels per month (New York)') +
        labs(x = "Month", y = "Travels per month")
```

```r
# Displaying few rows of Chicago dataset.
# Notice the las "Month" column.


head(chi)
```

```r
# Plotting month wise travels for Chicago.

ggplot(aes(x=as.numeric(Month)), data=chi) + 
        geom_histogram(binwidth=0.5) +
        scale_x_continuous(breaks=1:6) +
        ggtitle('Histogram of travels per month (Chicago)') +
        labs(x = "Month", y = "Travels per month")
```

```r
# Displaying few rows of Washington dataset.
# Notice the las "Month" column.


head(wash)
```

```r
# Plotting month wise travels for Washington.

ggplot(aes(x=as.numeric(Month)), data=wash) + 
        geom_histogram(binwidth=0.5) +
        scale_x_continuous(breaks=1:6) +
        ggtitle('Histogram of travels per month (Washington)') +
        labs(x = "Month", y = "Travels per month")
```

**Summary of your question 2 results goes here.**

As visible in the plots, the maximum number of travels started in the sixth month(June) in all the three cities.

### Question 3

**Your question 3 goes here.**

**Find the most common month of starting of travels gender wise.**


```r
# Plotting month wise travels for Chicago on genderly basis.

ggplot(aes(x=as.numeric(Month)), data=subset(chi, Gender %in% c("Male", "Female"))) + 
        geom_histogram(binwidth=0.5) + 
        # scale_x_continuous(limits=c(1930, 2010)) +
        facet_wrap(~Gender) +
        ggtitle('Histogram of travels per month genderwise (Chicago)') +
        labs(x = "Month", y = "Travels per month")
        
```

```r
# Plotting month wise travels for New York on genderly basis.

ggplot(aes(x=as.numeric(Month)), data=subset(ny, Gender %in% c("Male", "Female"))) + 
        geom_histogram(binwidth=0.5) + 
        # scale_x_continuous(limits=c(1930, 2010)) +
        facet_wrap(~Gender) +
        ggtitle('Histogram of travels per month genderwise (New York)') +
        labs(x = "Month", y = "Travels per month")
        
```

**Summary of your question 3 results goes here.**

1. In New York, most travels started in June for both Men and Women.
2. In Chicago, most travels started in June for both Men and Women.
